```python
from PIL import Image, ImageDraw
```

```python
orig_im = Image.open('../urchin-embryo.png')
print(orig_im.format, orig_im.size, orig_im.mode)

orig_im = orig_im.resize((512, 512))
print(orig_im.format, orig_im.size, orig_im.mode)
```

```python
orig_im
```

```python
im = orig_im.copy()
imdraw = ImageDraw.Draw(im)

macromere1 = [(120, 250),
              (220, 250),
              (220, 350),
              (120, 350) ]
macromere2 = [(320, 270),
              (420, 270),
              (420, 370),
              (320, 370) ]

mesomere1 = [(140, 100),
              (220, 100),
              (220, 180),
              (140, 180) ]
mesomere2 = [(300, 100),
              (380, 100),
              (380, 180),
              (300, 180) ]

micromere1 = [(195, 410),
              (210, 410),
              (210, 425),
              (195, 425) ]
micromere2 = [(250, 415),
              (265, 415),
              (265, 430),
              (250, 430) ]

imdraw.polygon(macromere1, fill='red')
imdraw.polygon(macromere2, fill='pink')
imdraw.polygon(mesomere1, fill='cyan')
imdraw.polygon(mesomere2, fill='red')

imdraw.polygon(micromere1, fill='green')
imdraw.polygon(micromere2, fill='orange')


im
```
